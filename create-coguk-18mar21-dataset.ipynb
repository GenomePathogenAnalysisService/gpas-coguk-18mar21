We can pull the `cogid` out of the `sequence_name`

```python
def split_sequence_name(row):
    cols=row['sequence_name'].split('/')
    if len(cols)==3:
        return(cols[1])
    else:
        return(None)

cog_metadata_mar18['cogid']=cog_metadata_mar18.apply(split_sequence_name,axis=1)    
cog_metadata_mar18[:3]
```

```python
batch0=pandas.read_csv('set0.csv',names=['FASTQ_FILENAME'])
batch1=pandas.read_csv('set1.csv',names=['FASTQ_FILENAME'])
batch2=pandas.read_csv('set2.csv',names=['FASTQ_FILENAME'])
```

```python
def create_ena_accession(row):
    stem=row.FASTQ_FILENAME.split('.fastq.gz')[0]
    if "_" in stem:
        return stem[:-2]
    else:
        return(stem)


batch0['ENA_ACCESSION']=batch0.apply(create_ena_accession,axis=1)
batch1['ENA_ACCESSION']=batch1.apply(create_ena_accession,axis=1)
batch2['ENA_ACCESSION']=batch2.apply(create_ena_accession,axis=1)
```

```python
set0=set(batch0['ENA_ACCESSION'].unique())
set1=set(batch1['ENA_ACCESSION'].unique())
set2=set(batch2['ENA_ACCESSION'].unique())

entire_set = set0 | set1 | set2

len(entire_set)
```

```python
def split_fastq_files(row):

   
    cols=row['fastq_ftp'].split(';')
    
    unpaired_reads=False
    
    fastq_files=[]
    
    for i in cols:
        foo=i.split('/')[-1].split('.fastq.gz')[0]

        bits=foo.split("_")
        
        uid=bits[0]
        
        if uid==foo:
            unpaired_reads=True
        
        
        fastq_files.append(foo)
        
    
    
    is_illumina=False
    if uid+"_1" in fastq_files and uid+"_2" in fastq_files:
        is_illumina=True
        
        
    return(pandas.Series([unpaired_reads,is_illumina,len(fastq_files)]))

df[['HAS_UNPAIRED_READ_FILE','ASSUMED_ILLUMINA','NUMBER_FASTQ_FILES']]=df.apply(split_fastq_files,axis=1)    
```
